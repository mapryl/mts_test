## 1. Перечислите алгоритмы для анализа временных рядов
### 1.1. Autoregression (AR)
Авторегрессия - это модель временных рядов, которая использует предыдущие наблюдения в качестве входных данных в уравнение регрессии для прогнозирования значения на следующем шаге.
### 1.2. Moving Average (MA)
Алгоритм скользящего среднего использует прошлые прогнозируемые ошибки (или шум) в регрессионной модели для определения усредненного тренда по данным. 
Этот метод предполагает, что компоненты тренда и сезонности временного ряда уже удалены или скорректированы.
### 1.3. Autoregressive Moving Average (ARMA)
Cочетает в себе модели авторегрессии (AR) и скользящего среднего (MA). По сравнению с чистыми моделями AR и MA, модели ARMA показывают большую эффективность в моделировании стационарных временных рядов. Модель находит следующий шаг в последовательности как линейную функцию наблюдений и ошибок на предыдущих временных шагах.
Метод подходит для одномерных временных рядов без трендовых и сезонных составляющих. 
### 1.4. Autoregressive Integrated Moving Average (ARIMA)
Сочетает в себе модели авторегрессии (AR) и скользящего среднего (MA), а также этап предварительной обработки, называемый интегрированием (I), чтобы сделать последовательность стационарной.  
Метод подходит для одномерных временных рядов с трендом и без сезонных составляющих.
### 1.5. Seasonal Autoregressive Integrated Moving-Average (SARIMA)
SARIMA моделирует следующий шаг в последовательности как линейную комбинацию наблюдений, ошибок, сезонных наблюдений и сезонных ошибок на предыдущих временных шагах.
Метод подходит для одномерных временных рядов с трендовой и / или сезонной составляющими. 
### 1.6. Seasonal Autoregressive Integrated Moving-Average with Exogenous Regressors (SARIMAX)
Интегрированное скользящее среднее сезонной авторегрессии с экзогенными регрессорами (SARIMAX) - это расширение модели SARIMA, которое также включает моделирование экзогенных переменных. Экзогенные переменные также называются ковариатами и могут рассматриваться как параллельные входные последовательности, которые имеют наблюдения на тех же временных шагах, что и исходный ряд. Первичная серия может называться эндогенными данными, чтобы отличать ее от экзогенной последовательности (последовательностей). Наблюдения за экзогенными переменными включаются в модель непосредственно на каждом временном шаге и не моделируются таким же образом, как первичная эндогенная последовательность.
Метод подходит для одномерных временных рядов с трендовыми и / или сезонными компонентами и экзогенными переменными.
### 1.7. Vector Autoregression (VAR)
Метод векторной авторегрессии (VAR) моделирует следующий шаг в каждом временном ряду с использованием модели AR. Это обобщение AR на несколько параллельных временных рядов, например, многомерный временной ряд.
### 1.8. Vector Autoregression Moving-Average (VARMA)
Метод векторной авторегрессии скользящего среднего (VARMA) моделирует следующий шаг в каждом временном ряду с использованием модели ARMA. Это обобщение ARMA на несколько параллельных временных рядов, например многомерный временной ряд. 
Метод подходит для многомерных временных рядов без трендовых и сезонных составляющих.
### 1.9. Vector Autoregression Moving-Average with Exogenous Regressors (VARMAX)
Скользящее среднее векторной авторегрессии с экзогенными регрессорами (VARMAX) - это расширение модели VARMA, которое также включает моделирование экзогенных переменных. 
Это многовариантная версия метода ARMAX. Метод подходит для многомерных временных рядов без тренда и сезонных компонентов с экзогенными переменными.

## 3. Что такое шестая нормальная форма? Целесообразно ли ее использование.
Таблица находится в шестой нормальной форме (6NF), если её строки содержат первичный ключ и не более одного атрибута. Из этого определения следует, что таблица находится в 6NF, когда она неприводима, то есть не может быть подвергнута дальнейшей декомпозиции без потерь. Стоит отметить, что таблица, которая находится в 6NF, также находится и в 5NF, и во всех предыдущих.
В некоторых случаях такая нормализация может быть полезна, так как она позволяет бороться с избыточностью данных, но в то же время данных нормализация таблиц до шестой нормальной формы может приводить к значительному снижению производительности. Кроме этого такая нормализация сделает работу с базой данных очень сложной за счет многократного увеличения количества таблиц, поэтому шестую нормальную форму в реальном мире используют достаточно редко.

## 4. Что такое расстояние Левенштейна и где оно используется?
Расстояние Левенштейна - это строковая метрика для измерения разницы между двумя последовательностями. Неформально, расстояние Левенштейна между двумя словами - это минимальное количество односимвольных правок (вставок, удалений или замен), необходимых для преобразования одного слова в другое. Может быть вычислено рекурсивно и итеративно (через динамическое программирование).
Используется для проверки орфографии (коррекции опечаток), распознавания текстов, помощи в машинном переводе.

```python
# будем исследовать зависимость средней стоимости квартиры в Лондоне от времени
hpi_to_research = hpi[hpi.RegionName == 'London'][['Date', 'FlatPrice']]
# удалим строчки с отсутствующей ценой
hpi_to_research.dropna(inplace=True)
```

```python
# приведем дату к нужному формату
hpi_to_research['Date'] = pd.to_datetime(hpi_to_research['Date'], dayfirst=True)
```

```python
hpi_to_research.head()
```

```python
hpi_to_research.plot(x='Date', y='FlatPrice', legend=False, figsize=(10,6))
```

```python
hpi_to_research.index = hpi_to_research.Date
hpi_to_research = hpi_to_research.asfreq('MS').drop(['Date'], axis=1)
```

```python
hpi_to_research.head()
```

```python
# уберем тренд для получения сезонной составляющей
decompose_result = seasonal_decompose(hpi_to_research[['FlatPrice']], model="multiplicative")
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,5))
decompose_result.trend.plot(ax=ax[0])
decompose_result.seasonal.plot(ax=ax[1])
plt.tight_layout()
```

```python
# для одного года отдельно (заметно, что стоимость вырастает летом)
decompose_result.seasonal[:12].plot(figsize=(12,2))
```

```python
def check_stationarity(column_to_test, name):
    test_stat = sm.tsa.adfuller(column_to_test)
    print(f'ряд {name} {"не " if test_stat[0] > test_stat[4]["5%"] else ""}стационарен')
```

```python
# проверим ряд на стационарность
check_stationarity(hpi_to_research['FlatPrice'], 'FlatPrice')
```

```python
# возьмем разность ряда
flat_price_diff = hpi_to_research['FlatPrice'].diff(periods=12).dropna()
fig,ax = plt.subplots(3,1,figsize=(10,8))
flat_price_diff.plot(ax=ax[0])
sm.graphics.tsa.plot_acf(flat_price_diff.diff().dropna(), lags=25, ax=ax[1])
sm.graphics.tsa.plot_pacf(flat_price_diff.diff().dropna(), lags=25, ax=ax[2])
plt.tight_layout()
```

```python
"""будем использовать модель SARIMA
порядок дифференцирования - 1 (в данных есть тренд)
порядок сезонного дифференцирования - 1 (данные не стационарные), 
порядок сезонной авторегрессии - 1 (используем данные за этот месяц в прошлом году),
порядок MA  - 1 (используем данные за последний год)"""
train, test = hpi_to_research['FlatPrice'][:-24], hpi_to_research['FlatPrice'][-24:]
s_mod = SARIMAX(train, 
                order=(1,1,0), 
                seasonal_order=(0,1,1,12), 
               trend='t').fit()
predictions = s_mod.predict()
```

```python
s_mod.summary()
```

```python
# посмотрим на график остатков и проверим их на стационарность
fig,ax = plt.subplots(3,1,figsize=(10, 7))
s_mod.resid.plot(ax=ax[0])
sm.graphics.tsa.plot_acf(s_mod.resid[24:], lags=25, ax=ax[1])
sm.graphics.tsa.plot_pacf(s_mod.resid[24:], lags=25, ax=ax[2])
print(f'Критерий Дики-Фуллера: p={sm.tsa.stattools.adfuller(s_mod.resid[24:])[1]:.4f}')
plt.tight_layout()
```

```python
# сделаем предсказания на последние два года
forecast = s_mod.forecast(24)
```

```python
# изобразим на графике предсказания за 4 года
fig, ax = plt.subplots(figsize=(12, 3))
train.append(forecast)[-48:].plot(ax=ax)
hpi_to_research[-48:].plot(y='FlatPrice', ax=ax, label='true')
```

```python
# оценим предсказания на train (предсказывается одно значение за раз), лучшее значение метрики = 1
print(f'SARIMA model R2 score (train): {r2_score(train, s_mod.predict())}')
```

```python
# оценим предсказания на train (предсказывается одно значение за раз), лучшее значение метрики = 1
print(f'SARIMA model R2 score (test): {r2_score(test, forecast)}')
# модель показывает достаточно плохие результаты в долгосрочном предсказании 
# (что можно объяснить сложной природой входных данных)
```
